```python
import urllib.request, json
```

```python
url = "https://raw.githubusercontent.com/jasonwang1031/DataViz_FinalData/master/manhattan.geojson"
response =urllib.request.urlopen(url)
```

```python
data = json.loads(response.read())
```

```python
data['features'][0]['geometry']['type']
```

```python
only_Line = list(filter(lambda x: x['geometry']['type'] != 'Point', data['features']))
```

```python
data
```

```python
new_data = {}
```

```python
new_data['type'] = data['type']
```

```python
new_data['crs'] = data['crs']
```

```python
new_data['features'] = only_Line
```

```python
new_data
```

```python
with open('manhattan_line.txt', 'w') as outfile:
    json.dump(new_data, outfile)
```

```python
only_Point = list(filter(lambda x: x['geometry']['type'] == 'Point', data['features']))
```

```python
new_data = {}
new_data['type'] = data['type']
new_data['crs'] = data['crs']
new_data['features'] = only_Point
```

```python
with open('manhattan_point.txt', 'w') as outfile:
    json.dump(new_data, outfile)
```
